#### 1. capitalize() – Converts the first character to uppercase and the rest to lowercase.

```python
name = 'raghavendar'
name.capitalize()
```

#### 2. casefold() – Converts the string to lowercase in a more aggressive way than lower().
Performs a more aggressive and Unicode-aware conversion to lowercase.
Designed for case-insensitive comparisons.
Handles special characters like German ß (sharp S):
* "ß".lower() → "ß"
* "ß".casefold() → "ss"

Useful when comparing strings across different languages or scripts.

```python
name = 'RAGHveßNDAR'
name.casefold()
```

#### 3. center(width[, fillchar]) – Returns a centered string of given width, padded with a specified character.


```python
name = '   RAGHveNDAR                      '
name.center(56)
```

#### 4. count(sub[, start[, end]]) – Returns the number of occurrences of a substring in the string.


```python
name = 'raghavendar'
name.count('r')
```

```python
name = 'Python is a very good course. It is useful for machine learning'
print(name.count("is"))
print(name.count("is",0,15))
```
