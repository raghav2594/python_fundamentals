# Stone Paper Scisor Game

### Simple use case

```python
user_turn = input("Stone Paper Scisor Shoot!!!!");
computer_turn = 'paper'
if user_turn.lower() == computer_turn.lower():
    print("Match Tie")
elif user_turn.lower() == 'stone' and computer_turn.lower() == 'paper':
    print("Computer Win")
elif user_turn.lower() == 'stone' and computer_turn.lower() == 'scisor':
    print("You Win")
elif user_turn.lower() == 'paper' and computer_turn.lower() == 'stone':
    print("You Win")    
elif user_turn.lower() == 'paper' and computer_turn.lower() == 'scisor':
    print("Computer Win")
elif user_turn.lower() == 'scisor' and computer_turn.lower() == 'stone':
    print("Computer Win")
elif user_turn.lower() == 'scisor' and computer_turn.lower() == 'paper':
    print("You Win")
else:
    print("Invalid Parameter")
```

```python
import random as random
game_parameter = ['stone', 'paper', 'scisor']
while True:
    random_index = random.randint(0,2)
    computer_turn = game_parameter[random_index]
    user_turn = input("Stone Paper Scisor Shoot!!!!");    
    if user_turn.lower() == computer_turn.lower():
        print("Match Tie")
    elif user_turn.lower() == 'stone' and computer_turn.lower() == 'paper':
        print("Computer Win")
    elif user_turn.lower() == 'stone' and computer_turn.lower() == 'scisor':
        print("You Win")
    elif user_turn.lower() == 'paper' and computer_turn.lower() == 'stone':
        print("You Win")    
    elif user_turn.lower() == 'paper' and computer_turn.lower() == 'scisor':
        print("Computer Win")
    elif user_turn.lower() == 'scisor' and computer_turn.lower() == 'stone':
        print("Computer Win")
    elif user_turn.lower() == 'scisor' and computer_turn.lower() == 'paper':
        print("You Win")
    else:
        print("Invalid Parameter")
    retry = input("Do you want to play again say!!")
    if retry.lower() != 'yes':
        print("Thanks for playing the game")
        break
    
```

### Improvisation

```python
import random as random
game_parameter = ['stone', 'paper', 'scisor']
while True:
    random_index = random.randint(0,2)
    computer_turn = game_parameter[random_index]
    user_turn = input("Stone Paper Scisor Shoot!!!!");
    print(f'User_Input : {user_turn}, Computer_Input : {computer_turn}')
    if user_turn.lower() == computer_turn.lower():
        print("Match Tie")
    elif (user_turn.lower() == 'stone' and computer_turn.lower() == 'paper') or (user_turn.lower() == 'paper' and computer_turn.lower() == 'scisor') or (user_turn.lower() == 'scisor' and computer_turn.lower() == 'stone'):
        print("Computer Win")
    elif (user_turn.lower() == 'stone' and computer_turn.lower() == 'scisor') or (user_turn.lower() == 'paper' and computer_turn.lower() == 'stone') or (user_turn.lower() == 'scisor' and computer_turn.lower() == 'paper'):
        print("You Win")
    else:
        print("Invalid Parameter")
    retry = input("Do you want to play again say!!")
    if retry.lower() != 'yes':
        print("Thanks for playing the game")
        break
```
