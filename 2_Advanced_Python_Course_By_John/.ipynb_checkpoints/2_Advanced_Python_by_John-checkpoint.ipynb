### Positional Arguments

```python
def get_details(name,place,expected_price):
    print("Client Name:",name)
    print("Place:",place)
    print("Expected Price:",expected_price)
```

```python
get_details("Raghav","Salem",97.4) #calling in actual function order
```

```python
get_details(place= "Salem",expected_price=97.4, name="Raghav") #calling in actual function order
```

### Default Parameters and Arguments
 * Default parameter should always placed at the last

```python
def get_details(name,expected_price,place="Chennai"):
    print("Client Name:",name)
    print("Place:",place)
    print("Expected Price:",expected_price)
```

```python
get_details(expected_price=97.4, name="Raghav") #calling the function uses default place parameter value
```

```python
get_details(expected_price=97.4, name="Raghav", place="Salem") #calling the function with dynamic place parameter
```

### Variable length argument
1. parameter with * can hold many arguments as Tuples
2. Positional argument calling is not allowed 

```python
def get_details(name,place, *expected_price):
    print("Client Name:",name)
    print("Place:",place)
    print("Expected Price:",expected_price)
```

```python
get_details("Raghav","Chennai",98.7,90,81)
```

### Keyword argument
1. parameter with ** can hold many arguments as key value pair like dictionary
2. Positional argument calling is allowed

```python
def get_details(name,place="Salem", **expected_price):
    print("Client Name:",name)
    print("Place:",place)
    print("Expected Price:",expected_price)
```

```python
get_details("Raghav",place="Chennai",key1=98.7,key2=90,key3=81)
```

## 5.4 Challenge to find the factors of the number

If the user enters 20, output should be "The factors of 20 is 1,2,4,5,10,20

```python
factor_number = int(input("Enter the number to find the factor of the number"))
result = []
for i in range(1,factor_number+1):
    if factor_number%i == 0:
        #print(i,end = ' ')
        result.append(i)
print("The factors of {} is {}".format(factor_number,result))
```
