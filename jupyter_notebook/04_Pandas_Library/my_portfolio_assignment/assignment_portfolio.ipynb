## Data Cleaning
* Removing the unwanted rows and first columns initially
* Set the Header row
* Remove unwanted column 

```python
portfolio_data = pd.read_excel("portfolio1.xlsx",header=22) # headers start at row 23
portfolio_data = portfolio_data.drop(portfolio_data.columns[0], axis=1)  # drop first column
portfolio_data.drop(columns=["Quantity Pledged (Margin)","Quantity Pledged (Loan)", "Quantity Discrepant","Unrealized P&L Pct."], inplace = True) #Drop unwanted columns
portfolio_data = pd.DataFrame(portfolio_data)
portfolio_data
```

```python
portfolio_data.shape
```

### Add 2 columns one for 'Invested' and another for 'Current Value'

```python
portfolio_data["Invested"] = portfolio_data["Quantity Available"] * portfolio_data["Average Price"]
portfolio_data["Current Value"] = portfolio_data["Quantity Available"] * portfolio_data["Previous Closing Price"]
portfolio_data
```

### Based on Sector how much money invested
Using GroupBy

```python
portfolio_data.groupby(by ="Sector")["Invested"].sum()
```

### Based on Sector and Stocks how much money invested
Using GroupBY

```python
portfolio_data.groupby(by =["Sector","Symbol"])["Invested"].sum()
```

### Based on Sector and how much percentage invested
Using GroupBY

```python
sector_sum = portfolio_data.groupby(by =["Sector", "Symbol"])["Invested"].sum()
sector_percentage = sector_sum/sector_sum.sum() * 100
sector_percentage = sector_percentage.round(2).astype("str") + '%'
pd.DataFrame({'Invested':sector_sum,
              'Percentage': sector_percentage})
```

```python
# Create DataFrame for sector summary + percentage 
sector_summary = pd.DataFrame({ 'Invested': sector_sum, 'Percentage': sector_percentage })
# Write everything into one Excel file with multiple sheets
with pd.ExcelWriter("portfolio_output.xlsx", engine="openpyxl") as writer: 
    portfolio_data.to_excel(writer, sheet_name="PortfolioData", index=False) 
    portfolio_data.groupby("Sector")["Invested"].sum().to_excel(writer, sheet_name="SectorTotal")
    portfolio_data.groupby(["Sector", "Symbol"])["Invested"].sum().to_excel(writer, sheet_name="SectorSymbolTotal") 
    sector_summary.to_excel(writer, sheet_name="SectorPercentage") 
print("All data written successfully to portfolio_output.xlsx")
```

## Do the Same above scenarios by using PivotTable

### Based on Sector how much money invested
Using PivotTable

```python
pd.pivot_table(data=portfolio_data,index='Sector',values="Invested",margins=True)
```

### Based on Sector and Stocks how much money invested

Using PivotTable

```python
pd.pivot_table(data=portfolio_data, values="Invested", index=["Sector", "Symbol"],margins='all')
```

### Based on Sector and how much percentage invested
Using PivotTable

```python
pivot_Invested = pd.pivot_table(data=portfolio_data, values="Invested",index=["Sector","Symbol"], aggfunc='sum')
pivot_Invested["percentage"] = pivot_Invested["Invested"] /pivot_Invested["Invested"].sum() * 100
pivot_Invested
```
