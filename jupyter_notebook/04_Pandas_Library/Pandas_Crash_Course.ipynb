## Task 1 : Find Out the top 5 selling items  individual revenue

```python
#order_details.groupby("item_name")["item_price"].sum().sort_values(ascending=False)
#order_details.groupby("item_name").first()
#My Solution
order_details.groupby("item_name").agg(
    quantity = ("quantity", "sum"),
    revenue = ("item_price", "sum")
).sort_values('quantity', ascending=False).head(5)

```

```python
#Copilot suggested
top5 = (order_details.groupby('item_name', as_index=False).agg(
          total_quantity=('quantity', 'sum'),
          total_revenue=('item_price', 'sum')  # item_price is line total per row
           ).sort_values('total_quantity', ascending=False).head(5))

top5['total_revenue'] = top5['total_revenue'].map(lambda x: f"${x:,.2f}")
top5
```

## Task 2 : Find Out the top 5 selling items  total revenue

```python
top5 = order_details.groupby("item_name", as_index=False).agg(
    quantity = ("quantity", "sum"),
    revenue = ("item_price", "sum")
).sort_values('quantity', ascending=False).head(5)
top5.agg(
    total_revenue = ("revenue","sum")
)
```
