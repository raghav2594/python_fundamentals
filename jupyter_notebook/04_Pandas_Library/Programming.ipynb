### Create a list of numbers and print only even numbers using a forloop

```python
numbers = range(0,20)
even_number = []
for i in numbers:
    if i%2 == 0:
        even_number.append(i)
even_number
```

### Write a program that handles division by zero using try-except?

```python
try:
    num = int(input("Enter Numerator"))
    den = int(input("Enter denominator"))
    result = num/den
    print(result)
except ZeroDivisionError:
    print("Denominator should not be zero")
except ValueError:
    print("Values always should be numbers")
finally:
    print("Program Endded")
```

### Given a String, count how many vowels it has

```python
statement = "This is the test statement"
count = 0
for i in statement:
    if i in "aeiou":
        #count = count + 1
        count += 1
print("Total Vowels are", count)

```
