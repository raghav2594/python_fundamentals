#### 2. Sets

- 1 Dimension
- Hetrogenous
- It won't allow duplicate values
- Indexing and Slicing is not possible
- Mutable

```python
empty_set = set()
```

```python
empty_set
```

```python
items_to_buy_2 = {"Apple", "Orange", "Grapes", "Pomogranate", "Apple", 6.75, 8, True}
```

```python
items_to_buy_2
```

```python
items_to_buy_2.add("Bananna")
items_to_buy_2
```

#### 3 Tuple

- 1 Dimension
- Hetrogenous

- Slicing is possible
- Immutable




```python
empty_tuple = ()
```

```python
type(empty_tuple)
```

```python
items_to_buy_3 = (7,8, "Apple", "Orange", "Grapes", "Pomogranate", "Apple", 6.75, 8, True)
items_to_buy_3
```
