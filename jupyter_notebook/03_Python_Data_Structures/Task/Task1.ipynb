# Example for Python Datastructure

## List

```python
list_of_items = ["Pen", "Paper", "Pencil", "Rubber", "HeadPhone", "Mike", "Mouse", "Paper", "Monitor", "Laptop" , "Keyboard"]
```

```python
list_of_items.append("Remote") #Appends the items at last
list_of_items
```

```python
list_of_items.clear() #Clear all the items inside the list
list_of_items
```

```python
list_of_items.copy()
copy_item = list_of_items  # both refers to the same reference
copy_item
```

```python
list_of_items.append("car")
copy_item
```

```python
copy_item = list_of_items.copy() #This is having separate reference
list_of_items.append("Jeep")
copy_item
```

```python
list_of_items.count("Paper")
```

```python
list_of_items.extend(copy_item) #Iterate copy_item and stores inside the list_of_items
list_of_items
```

```python
list_of_items.index("car")
```

```python
list_of_items.pop(1) #Remove items from the list based on Index
list_of_items
```

```python
list_of_items.remove("Paper") #First Occurence would be removed
list_of_items
```

```python
list_of_items.reverse() #Reversing the order of the list
list_of_items
```

```python
list_of_items.sort() #Default ascendig order
list_of_items
```

```python
list_of_items.sort(reverse=True) #Default ascendig order
list_of_items
```
