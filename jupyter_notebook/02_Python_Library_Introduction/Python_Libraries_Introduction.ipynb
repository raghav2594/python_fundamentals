### 2. Numpy Library
- For Math and Array operations
- Basically playing with numbers
The NumPy library is installed automatically when you install Pandas.

```python
import numpy as np
```

```python
employee_repeated_data = np.repeat(a = employee_data,repeats = 2, axis= 0)
# Arguments information
# a - it can be a dataset might be an array or pandas dataset
# repeats - how many times want to repeat?
# axis - It should be repeated in column or rows
   # axis = 0 for repeating in rows
   # axis = 1 for repeating in columns
```

```python
pd.DataFrame(data = employee_repeated_data)
#A Pandas DataFrame is a two-dimensional, tabular data structure in Python provided by the Pandas library.
#It’s similar to a spreadsheet or SQL table and is one of the most powerful tools for data analysis.
```
