# Number Guessing Game

###  Level 1. Print if the user guessed number is right or wrong

```python
secret_number = 50

user_input = int(input("Enter the guessing number"))
#print(f'User Guessed number is {user_input}')

if secret_number == user_input:
    print("Congratulations your guess is right")
else:
    print("Sorry. Your Guess is wrong")
```

###  Level 2. Print if the user guessed number is right or wrong using Loop

#### 2.1 Using For loop - end iteration number is defined
Here the end iteration number 7 which means Maximum of 7 trails

```python
secret_number = 50
for i in range(0,7):
    user_input = int(input("Enter the guessing number"))
    if secret_number == user_input:
        print("Congratulations your guess is right")
        break
    else:
        print("Sorry. Your Guess is wrong")
```

#### 2.2 While loop - when end number is not defined and instructions has to executed the infinite times

```python
secret_number = 50
while True:
    user_input = int(input("Enter the guessing number"))
    if secret_number == user_input:
        print("Congratulations your guess is right")
        break
    else:
        print("Sorry. Your Guess is wrong")
```

###  Level 3. Give Multiple conditions

1. User should Enter the number between 1-100. If it's entered wrong, the program should say, "Invalid Number"
2. If the user guessed number > secret number, then program should say, "Your number is greater than the secret number"
3. If the user guessed number < secret number, then program should say, "Your number is lesser than the secret number"

```python
print("Welcome to Number Guessing Game")
print("Guess the number between 1-100")

secret_number = 98

for i in range(1,8):
    user_input = int(input("Enter the Number"))
    if user_input < 1 or user_input > 100:
        print("Invalid Number! Please enter the number between 1-100")
    elif user_input > secret_number:
        print("Your number is Higher")
    elif user_input < secret_number:
        print("Your number is Lesser")
    else:
        print("Your guess is right")
        print(f"Congratulations, you have guessed the number {secret_number} in {i} attempts" )
        break   
```

### Level 6 : Error Handling
If the user enters invalid characters other than numbers, errors are currently displayed. Instead, the error should be handled gracefully.

```python
import random

print("Welcome to Number Guessing Game")
print("Guess the number between 1-100")

secret_number = random.randint(1,100)
for i in range(1,8):
    try:
        user_input = int(input("Enter the Number"))
        if user_input < 1 or user_input > 100:
            print("Invalid Number! Please enter the number between 1-100")
        elif user_input > secret_number:
            print("Your number is Higher")
        elif user_input < secret_number:
            print("Your number is Lesser")
        else:
            print("Your guess is right")
            break
    except ValueError:
        print("Invalid character.Please enter the number between 1-100")

if secret_number == user_input:
   print(f"Congratulations, you have guessed the number {secret_number} in {i} attempts" )
else:
    print("You have exceeded the number of attempt to guess the number. Please try again")
```
