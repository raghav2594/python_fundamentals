## On what basis should I choose between charts?

1. Univariate Analysis  - To understand the quality of one feature(Column) - Ex: Countplot, Histogram, Boxplot, PieChart
2. Bivariate Analysis  - To understand the quality/relationship of two features(Column)  - Ex: BarPlot, Scatterplot
3. Mulrivariate Analysis - To understand the quality/relationship of more than two features(Column) - Ex: LinePlot,

## 3.2 Perform Data Visualization for better understanding

### 3.2.1 Univariate Analysis

```python
plt.hist(motor_data["mpg"]) # Speaks about the distribution of one continuous data
plt.xlabel("hp")
plt.ylabel("Frequency")
plt.title("MPG Distribution")
plt.show()
```

## 2 Types of data types

1. Discrete - Countable/ can't be measure/ can't be seggregated
2. Continuous - It has some units/ can be measurable / can be seggregated

### 3.2.2 Bivariate analysis

Using Barplot

```python
motor_data.head()
```

```python
# Barchart always talks about
plt.bar(x= motor_data["cyl"],height= motor_data["mpg"])
```

```python
import seaborn as sns
#Seaborn is a constructor of matplotlib
```

```python
sns.barplot(data = motor_data,x= "cyl", y = "mpg") # X label should be 'discrete', y - 'continuous'
plt.title("Cylinder vs MPG")
plt.show()      

# Understanding from the graph:
# Cars with 4 cylinders give around 25 mpg. Similarly, 6-cylinder cars give around 20 mpg,
# and 8-cylinder cars give around 15 mpg.
# Overall, as the number of cylinders increases, the mileage decreases.
# This is the interpretation of the diagram.
```
